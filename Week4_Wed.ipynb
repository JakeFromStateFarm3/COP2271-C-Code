* Alexander is walking from Tampa Downtown to Orlando Downtown (Assume that distance is 100 meters)
* Alexander is walking on straight line
* As he gets closer to the destination, he slows down

Based on his friend Aristotle's suggestion, he adjusts his walking speed (m/s) according to the formula

\begin{equation}
\Large v(t) = 0.5 {d(t) \over 1+0.2 ~ t}
\end{equation}

where $v(t)$ is his walking speed (m/s), and $d(t)$ is the remaining distance (m).

Hint: If the initial distance is 100 meters, speed will be 50 m/s and after 0.1 second, distance will reduce to 100-0.1*50 = 95 meters. Continue like this.

Instructions:

(1) Generate a plot showing the remaining distance $d(t)$ versus $t$, and find when Alexander will be within 3 meters of the final destination.

(2) Now, reduce your time step from 0.1 sec to 0.01 sec, and solve the problem again. Compare and comment.



```python
import math
import matplotlib.pyplot as plt
```

```python
# Initial distance
d = 100
t = 0

print(t, d)
```

```python
# After 0.1 second
d = d - 0.1 * (0.5*d) / (1+0.2*t)
t = t + 0.1

print(t, d)
```

```python
# After another 0.1 second
d = d - 0.1 * (0.5*d) / (1+0.2*t)
t = t + 0.1

print(t, d)
```

```python
tvals=[]
dvals=[]

t, d = 0, 100
tvals.append(t)
dvals.append(d)

for k in range(200):
  d = d - 0.1 * (0.5*d) / (1+0.2*t)
  t = t + 0.1
  tvals.append(t)
  dvals.append(d)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(tvals, dvals,'-')
plt.xlabel('time (s)')
plt.ylabel('Distance (m)')
plt.grid(True)
plt.title('Distance versus time graph')
#plt.ylim([0,5])
```
